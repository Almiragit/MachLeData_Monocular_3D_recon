```python
import wandb

# 1. Initialize a dummy run
# This will force W&B to identify the user and create a link
try:
    run = wandb.init(project="account-check")
    
    # 2. Print user info
    print("\n--- CONNECTION SUCCESSFUL ---")
    print(f"User: {run.entity}")  # This is your account/username
    print(f"Project: {run.project}")
    print(f"Run URL: {run.url}")
    print("-----------------------------\n")
    
    # 3. Log a simple test metric
    wandb.log({"status": "connected", "test_value": 1.0})
    
    # 4. Finish the run
    run.finish()

except Exception as e:
    print(f"Error: {e}")
    print("If you see an API key error, run '!wandb login --relogin' in a cell.")
```
