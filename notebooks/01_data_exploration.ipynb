```python
import sys
import os
import torch
import matplotlib.pyplot as plt
from PIL import Image

# 1. Add the project root to the Python path so we can import our src folder
sys.path.append(os.path.abspath(os.path.join('..')))

# 2. Import your custom classes
from src.data.dataset import MonocularDepthDataset
from src.models.model import DepthEstimationModel

# 3. Initialize the Dataset
try:
    # Make sure you have at least one image in data/raw/
    dataset = MonocularDepthDataset(stage="train")
    print(f"Dataset loaded successfully. Total images: {len(dataset)}")
    
    # Get the first sample
    image, depth = dataset[0]
    print(f"Image tensor shape: {image.shape}") # Expected: [3, 224, 224]
    print(f"Depth tensor shape: {depth.shape}") # Expected: [1, 224, 224]

    # 4. Visualization
    fig, ax = plt.subplots(1, 2, figsize=(12, 5))
    
    # Display Input Image (Need to permute from [C, H, W] to [H, W, C])
    ax[0].imshow(image.permute(1, 2, 0))
    ax[0].set_title("Input RGB Image")
    ax[0].axis('off')
    
    # Display Dummy Depth Map
    ax[1].imshow(depth.squeeze(), cmap='magma')
    ax[1].set_title("Initial Depth Map (Placeholder)")
    ax[1].axis('off')
    
    plt.show()

except Exception as e:
    print(f" Error: {e}")
    print("Hint: Check if your 'data/raw' folder has any .jpg or .png files.")

# 5. Sanity Check: Test the Model
model = DepthEstimationModel()
with torch.no_grad():
    test_output = model(image.unsqueeze(0)) # Add batch dimension
    print(f"Model forward pass successful. Output shape: {test_output.shape}")
```
