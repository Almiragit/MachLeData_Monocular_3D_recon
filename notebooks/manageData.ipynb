```python
import os
import shutil

# 1. Setup Paths
current_dir = os.getcwd()
if os.path.basename(current_dir) == 'notebooks':
    BASE_DIR = os.path.dirname(current_dir)
else:
    BASE_DIR = current_dir

SOURCE_DIR = os.path.join(BASE_DIR, 'src', 'data', 'request_data')
IMAGE_DEST = os.path.join(BASE_DIR, 'src', 'data', 'imageData')
LIDAR_DEST = os.path.join(BASE_DIR, 'src', 'data', 'lidarData')

# 2. Ensure Folders Exist
for folder in [SOURCE_DIR, IMAGE_DEST, LIDAR_DEST]:
    os.makedirs(folder, exist_ok=True)

def organize_files():
    # List all files in the folder
    try:
        files = os.listdir(SOURCE_DIR)
    except FileNotFoundError:
        print(f"Directory not found: {SOURCE_DIR}")
        return
    
    # 3. Find ALL image files (looking for '_image' with .jpg or .png)
    all_images = [f for f in files if '_image' in f.lower() and f.lower().endswith(('.jpg', '.jpeg', '.png'))]
    
    # 4. Limit to the first 100 images
    target_images = all_images[:100]
    
    if not target_images:
        print(f"No valid images found in {SOURCE_DIR}.")
        return

    print(f"Found {len(target_images)} images. Starting move...")
    moved_count = 0

    # 5. Loop through the list of target images
    for target_image in target_images:
        # Identify the unique ID (everything before '_image')
        unique_id = target_image.split('_image')[0]
        
        # Look for the matching depth file specifically for THIS unique_id
        target_depth = next((f for f in files if f.startswith(unique_id) 
                             and '_depth' in f.lower() 
                             and f.lower().endswith(('.tiff', '.tif'))), None)

        if target_depth:
            # Define paths
            img_src = os.path.join(SOURCE_DIR, target_image)
            img_dst = os.path.join(IMAGE_DEST, target_image)
            
            depth_src = os.path.join(SOURCE_DIR, target_depth)
            depth_dst = os.path.join(LIDAR_DEST, target_depth)

            # Move them
            shutil.move(img_src, img_dst)
            shutil.move(depth_src, depth_dst)
            
            moved_count += 1
            print(f"[{moved_count}] Success: Moved {unique_id}")
        else:
            print(f"⚠️ Warning: Found image {target_image}, but no matching .tiff file found for ID: {unique_id}")

    print(f"\nFinished! Total pairs moved: {moved_count}")

# Execute
organize_files()
```
