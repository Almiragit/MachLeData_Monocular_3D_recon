```python
import os
import shutil
from pathlib import Path

# 1. Setup paths
source_dir = Path('request_data')
image_dest = Path('imageData')
lidar_dest = Path('lidarData')

# 2. Create destination folders if they don't exist
image_dest.mkdir(exist_ok=True)
lidar_dest.mkdir(exist_ok=True)

def organize_first_pair():
    # Find all files ending with '_image.png'
    image_files = list(source_dir.glob("*_image.png"))
    
    if not image_files:
        print("No image files found ending with '_image.png'")
        return

    # Take only the first image found
    target_image = image_files[0]
    
    # Construct the expected depth filename based on the image's unique ID
    # This replaces '_image.png' with '_depth.tiff'
    base_name = target_image.name.replace('_image.png', '')
    target_depth = source_dir / f"{base_name}_depth.tiff"

    # 3. Check if the matching depth file exists
    if target_depth.exists():
        # Move the PNG to imageData
        shutil.move(str(target_image), str(image_dest / target_image.name))
        
        # Move the TIFF to lidarData
        shutil.move(str(target_depth), str(lidar_dest / target_depth.name))
        
        print(f"Successfully moved pair:")
        print(f" - {target_image.name} -> imageData/")
        print(f" - {target_depth.name} -> lidarData/")
    else:
        print(f"Found image {target_image.name}, but no matching depth file was found.")

# Run the function
organize_first_pair()
```
